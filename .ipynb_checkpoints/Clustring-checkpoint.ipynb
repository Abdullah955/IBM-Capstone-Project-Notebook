### Create function that foursquare API to retrive the data for each Neighborhood:

```python
from collections import Counter

def get_ven(Pos,lat,lng):
    
    ven_dict = {}
    ven_list = []
    
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            1000, 
            LIMIT)
    results = requests.get(url).json()
    try:
        re = results["response"]['groups'][0]['items'] 
    except:
        pass 
        
    for index, item in enumerate(re):   
        ven = results["response"]['groups'][0]['items'][index]['venue']['categories'][0]['icon']['prefix'].split('/')[-2]
        ven_list.append(ven)
    
    ven_dict[Pos] = dict(Counter(ven_list))
    
    return ven_dict
```

```python
dict_pos = {}

for index, row in df.iterrows():
    Postalcode = row['Postalcode']
    Latitude = row['Latitude']
    Longitude = row['Longitude']
    print(get_ven(Postalcode,Latitude,Longitude)[Postalcode])
    dict_pos[Postalcode] = get_ven(Postalcode,Latitude,Longitude)[Postalcode]
```

```python
df2 = pd.DataFrame.from_dict(dict_pos,orient='index',
                        columns=['food', 'shops', 'parks_outdoors', 'building','nightlife','travel','arts_entertainment','education'])
```

### Check number of cluster we need to do:

```python
# Detrimaine number of clusters
import numpy as np
from scipy.spatial.distance import cdist

# k means determine k
distortions = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(clus)
    kmeanModel.fit(clus)
    distortions.append(sum(np.min(cdist(clus, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / clus.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

### Cluster our data for 7 clusters:

```python
kmeans = KMeans(n_clusters=7).fit(clus)
```

```python
right = kmeans.labels_
```

```python
df = pd.DataFrame(right)
```
